```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset from sklearn
breast_cancer_data = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_data)
```

```python
#load the array into dataframe
breast_cancer_data_df = pd.DataFrame(breast_cancer_data.data,columns= breast_cancer_data.feature_names)
```

```python
#first 5 rows of the dataset
breast_cancer_data_df.head()
```

```python
#mearge the label into the dataframe
breast_cancer_data_df['label'] = breast_cancer_data.target
```

```python
#last 5rows of the dataset
breast_cancer_data_df.tail()
```

```python
#Assign into another dataframe
cancer_data = breast_cancer_data_df
```

```python
#Size of the dataset
cancer_data.shape
```

```python
#Getting some information about the data
cancer_data.info()
```

```python
#Checking the missing values
cancer_data.isnull().sum()
```

```python
#statistical measures
cancer_data.describe()
```

```python
#value count of label
cancer_data['label'].value_counts()
```

0 -----> malignant

1 -----> benign

```python
#Check the mean for label
cancer_data.groupby('label').mean()
```

```python
#X->Independent features , Y-Dependent features
X = cancer_data.drop('label',axis=1)
Y = cancer_data['label']
```

```python
#split the dataset into training and testing dataset
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,random_state=1,test_size=0.2)
```

```python
#Shape of X,X_train,X_test
print(X.shape,X_train.shape,X_test.shape)
```

```python
#Shape of Y.shape,Y_test.shape,Y_train.shape
print(Y.shape,Y_test.shape,Y_train.shape)
```

```python
#standardization of X_train
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
```

```python
#standardization of X_test
X_test = scaler.transform(X_test)
```

```python
#Loading the model and train it
model = LogisticRegression()
model.fit(X_train,Y_train)
```

```python
#predict the X_test and store it in Y_pred
Y_pred = model.predict(X_test)
```

```python
#accuracy score  of the model
accuracy = accuracy_score(Y_pred,Y_test)
print(accuracy)
```

```python
#final model
final_model = LogisticRegression()
final_model.fit(X,Y)
```
