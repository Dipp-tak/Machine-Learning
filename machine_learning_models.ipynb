#**RANDOM FOREST CLASSIFIER**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/random_forest_data.csv")
```

```python
data.head()
```

```python
NAs = pd.concat([data.isnull().sum()],axis=1,keys="data")
print(NAs)
```

```python
NAs[NAs.sum(axis=1) > 0]
```

```python
print(data["Age"].skew())
```

```python
plt.figure(figsize=(16,5))
sns.histplot(data["Age"],kde=True)
plt.show()
```

```python
data['Age']=data['Age'].fillna(data['Age'].mean())
```

```python
data['Cabin']=data['Cabin'].fillna(data['Cabin'].mode())
```

```python
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode())
```

```python
data['Pclass']=data['Pclass'].apply(str)
```

```python
for col in data.dtypes[data.dtypes=='object'].index:
  for_dummy = data.pop(col)
  data = pd.concat([data,pd.get_dummies(for_dummy,prefix=col)],axis=1)

data.head()
```

```python
def transform_true_false(data):
  return(data.astype(int))
```

```python
transformeddata=transform_true_false(data)
```

```python
data = transformeddata
data.head()
```

```python
X = data.drop('Survived',axis=1)
```

```python
Y = data['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y ,random_state = 1,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc_model = RandomForestClassifier(n_estimators=32)
rfc_model.fit(X_train,Y_train)
```

```python
Y_pred = rfc_model.predict(X_test)
```

```python
from sklearn.metrics import auc,roc_curve
false_positive_rate,true_positive_rate,thresolds= roc_curve(Y_test,Y_pred)
roc_auc = auc(false_positive_rate,true_positive_rate)
roc_auc
```

```python
n_estimators = [1,2,4,8,16,32,64,100,128,200]
train_result = []
test_result = []
for estimator in n_estimators:
  rfc_model = RandomForestClassifier(n_estimators = estimator,n_jobs=-1)
  rfc_model.fit(X_train,Y_train)
  train_pred = rfc_model.predict(X_train)
  false_positive_rate,true_positive_rate,thresolds= roc_curve(Y_train,train_pred)
  roc_auc = auc(false_positive_rate,true_positive_rate)
  train_result.append(roc_auc)
  Y_pred = rfc_model.predict(X_test)
  false_positive_rate,true_positive_rate,thresolds= roc_curve(Y_test,Y_pred)
  roc_auc = auc(false_positive_rate,true_positive_rate)
  test_result.append(roc_auc)

from matplotlib.legend_handler import HandlerLine2D
line1, = plt.plot(n_estimators,train_result,'b',label='Train_Auc')
line2, = plt.plot(n_estimators,test_result,'r',label='Test_Auc')
plt.legend(handler_map={line1:HandlerLine2D(numpoints=2)})
plt.ylabel("AUC_score")
plt.xlabel("n_estimators")
plt.show()
```

# **K-MEANS CLUSTRING**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
iris_data = sns.load_dataset('iris')
iris_data.head()
```

```python
from sklearn.cluster import KMeans
kmeans_model = KMeans(n_clusters=3)
```

```python
Y_pred = kmeans_model.fit_predict(iris_data.drop("species",axis=1))
```

```python
iris_data['Clust']=Y_pred
```

```python
iris_data.head()
```

```python
df1 = iris_data[iris_data.Clust==0]
df2 = iris_data[iris_data.Clust==1]
df3 = iris_data[iris_data.Clust==2]

plt.figure(figsize=(8, 6))
plt.scatter(df1['sepal_length'], df1['sepal_width'], color='red', label='Cluster 0')
plt.scatter(df2['sepal_length'], df2['sepal_width'], color='green', label='Cluster 1')
plt.scatter(df3['sepal_length'], df3['sepal_width'], color='blue', label='Cluster 2')


plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('K-Means Clustering of Iris Data')
plt.legend()
plt.show()
```

```python
sns.scatterplot(x='petal_length',y='petal_width',hue='species',data=iris_data)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df1['petal_length'], df1['petal_width'], color='red', label='Cluster 0')
plt.scatter(df2['petal_length'], df2['petal_width'], color='green', label='Cluster 1')
plt.scatter(df3['petal_length'], df3['petal_width'], color='blue', label='Cluster 2')


plt.xlabel('petal Length')
plt.ylabel('petal Width')
plt.title('K-Means Clustering of Iris Data')
plt.legend()
plt.show()
```

```python
k_range = range(1,20)
sse=[]
for k in k_range:
  kmeans_model = KMeans(n_clusters=k)
  kmeans_model.fit_predict(iris_data.drop('species',axis=1))
  sse.append(kmeans_model.inertia_)
```

```python
sse
```

```python
plt.plot(k_range,sse)
plt.show()
```
